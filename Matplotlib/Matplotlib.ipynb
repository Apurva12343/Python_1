```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('This is my x data')
plt.ylabel('This is my y Data')
plt.title('This is my Title')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)
plt.scatter(x,y,c=colors,s=sizes,alpha=.5)
plt.xlabel('this is my x Label')
plt.ylabel('this is my y label')
plt.title('Random Data')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)
plt.scatter(x,y,c='b',s=sizes,alpha=.5,marker='v')
plt.xlabel('this is my x Label')
plt.ylabel('this is my y label')
plt.title('Random Data')
```
