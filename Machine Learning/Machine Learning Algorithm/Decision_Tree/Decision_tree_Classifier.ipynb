```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
#Independant And dependent variables
x = df.iloc[:,:-1]
y = dataset.target
```

```python
x,y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

## Post Pruning


```python
##Post Pruning
tree_classifier = DecisionTreeClassifier()
```

```python
tree_classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(5,7))
tree.plot_tree(tree_classifier,filled=True)
```

```python
##post pruning with max depth 2
```

```python
tree_classifier = DecisionTreeClassifier(max_depth=2)
tree_classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(5,7))
tree.plot_tree(tree_classifier,filled=True)
```

```python
##Prediction
ypred = tree_classifier.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(ypred,y_test)
print(score)
```

```python
print(classification_report(ypred,y_test))
```

## PrePruning And Hyperparameter Tuning

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
parameter = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classfier = DecisionTreeClassifier()
clf = GridSearchCV(classfier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
## Train the data
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
classfier = DecisionTreeClassifier(criterion='log_loss',max_depth=4,max_features='sqrt',splitter='best')
```

```python
classfier.fit(x_train,y_train)
```

```python
y_pred = classfier.predict(x_test)
```

```python
accuracy_score(y_pred=y_pred,y_true=y_test)
```

```python
from sklearn import tree
plt.figure(figsize=(7,10))
tree.plot_tree(classfier,filled=True)
```
