# Anomaly/Outlier Dectection - ISolation Forest

```python
import pandas as pd
df = pd.read_csv('healthcare.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.14)
```

```python
clf.fit(df)
```

```python
predictions = clf.predict(df)
```

```python
predictions
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=predictions)
```

```python
import numpy as np
```

```python
index = np.where(predictions<0)
index
```

```python
x = df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors='r')
```
