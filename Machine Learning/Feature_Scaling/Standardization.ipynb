```python
##Standardization / z-score Normalization
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
```

```python
mean = np.mean(total_bill)
```

```python
std = np.std(total_bill)
```

```python
#Use Formula for normalization z-score = bill-mean/std
normalized_bill = []
for i in total_bill:
    z_score = (i-mean)/std
    normalized_bill.append(z_score)
```

```python
normalized_bill
```

```python
sns.histplot(total_bill,kde=True)
```

```python
sns.histplot(normalized_bill,kde=True)
```

```python
max(normalized_bill)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd
df1 = pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
sns.histplot(df1,kde=True)
```
