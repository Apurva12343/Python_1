```python
#5 Number Summary
# 1- minimum value
# 2 - Q1 - 25 percentile
# 3 - Median
# 4 - q3 - 75 percentile
#5 - maximum
```

```python
import numpy as np
marks = [45,66,45,78,99,5,47,87,89,23,25,1000,1100]
```

```python
# lower fence<----->higher fence
```

```python
minimum,q1,q2,q3,maximum = np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
iqr = q3-q1
```

```python
print(iqr)
```

```python
lower_fence = q1-1.5 * (iqr)
higher_fence = q3+1.5 * (iqr)
```

```python
lower_fence
```

```python
higher_fence
```

```python
outliers = []
for i in marks:
    if i >= lower_fence and i<=higher_fence:
        print("This Element is not an outlier")
    else:
        outliers.append(i)
```
