1 - Linear Interpolation

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=x,y=y)
```

```python
x_new = np.linspace(1,5,10) #creates new x values
y_interpol = np.interp(x_new,x,y) #interpolate y value
```

```python
sns.scatterplot(x=x_new,y = y_interpol)
```

2 - cubic interpolation with scipy

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
# it will create a cubic interpolation
f = interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interpol = f(x_new)
```

```python
print(y_interpol)
```

```python
sns.scatterplot(x = x_new,y = y_interpol)
```

3 - Polynomial Interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
#interpolate the data using polynomial interpolation
p = np.polyfit(x,y,4) # fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,10)
y_interpol = np.polyval(p,x_new)
```

```python
y_interpol
```

```python
sns.scatterplot(x=x_new,y=y_interpol)
```
